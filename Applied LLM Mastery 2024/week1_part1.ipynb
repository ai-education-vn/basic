# [Week 1, Part 1] Applied LLM Foundations and Real World Use Cases

## History and Background

![image](./img/overall_ai.png)

AI là một nhánh của khoa học máy tính liên quan đến việc tạo ra những cỗ máy có suy nghĩ và hành vi giống con người. ML là một lĩnh vực thuộc AI cho phép máy tính tìm hiểu các mẫu từ dữ liệu và đưa vào dự đoán mà không cần lập trình rõ ràng. NN (mạng thần kinh) là một tập hợp con của ML, mô phỏng cấu trúc của bộ não con người và rất quan trọng trong các thuật toán học sâu. DL một tập hợp con của NN, có hiệu quả trong việc giải quyết các vấn đề phức tạp, như đã thấy trong các công nghệ nhận dạng hình ảnh và dịch ngôn ngữ. GenAI là một tập con của DL có thể tạo ra nội dung dựa trên các mẫu đã học. LLM là một dạng của GenAI có mục đích tạo văn bản giống con người bằng cách học  từ dữ liệu văn bản mở rộng

GenAI và LLM đã cách mạng hóa lĩnh vực trí tuệ nhân tạo cho phép máy móc tạo ra nội dung đa dạng như văn bản, hình ảnh, nhạc, âm thanh và video. Các mô hình AI tổng quát tạo ra nội dung mới bằng cách học các mẫu và mối quan hệ từ các bộ dữ liệu do con người tạo ra.

Cốt lõi của AI tổng quát là các mô hình nền tảng về cơ bản đề cập đến các mô hình AI lớn có khả năng đa tác vụ, thực hiện các nhiệm vụ như tóm tắt, hỏi đáp và phân loại ngay lập tức. Những mô hình này, giống như mô hình phổ biến mà mọi người đã nghe nói đến ChatGPT, có thể thích ứng với các trường hợp cụ thể mà không cần đào tạo nhiều và tạo nội dung với dữ liệu mẫu tối thiểu.

Việc đào tạo AI tổng quát thường liên quan đến việc học có giám sát, trong đó mô hình được cung cấp nội dung do con người tạo ra và các nhãn tương ứng. Bằng cách học từ dữ liệu này, mô hình trở nên thành thạo trong việc tạo nội dung tương tự như tập huấn luyện.

GenAI không phải là một khái nhiệm mới. Một ví dụ đáng chú ý về GenAI là chuỗi Markov, một mô hình thống kê được nhà toán học người Nga Andrey Markov giới thiệu vào năm 1906. Các mô hình Markov ban đầu được sử dụng cho các nhiệm vụ như dự đoán từ tiếp theo nhưng tính đơn giản của chúng đã hạn chế khả năng tạo ra văn bản hợp lý.

Bối cảnh đã thay đổi đáng kể qua nhiều năm với sự ra đời của các kiến trúc mạnh mẽ hơn và bộ dữ liệu lớn hơn. Vào năm 2014, mạng GAN (generative adversarial networks) đã xuất hiện, sử dụng hai mô hình làm việc cùng nhau - một mô hình tạo ra đầu ra và mô hình còn lại phân biệt đối xử dữ liệu thực từ đầu ra được tạo ra. Cách tiếp cận này, được minh họa bằng các mô hình như StyleGAN, đã cải thiện đáng kể tính chân thực của nội dung được tạo ra.

Một năm sau, các mô hình phổ biến đã được giới thiệu, tinh chỉnh kết quả đầu ra của chúng nhiều lần để tạo ra các mẫu dữ liệu mới giống với tập dữ liệu huấn luyện. Sự đổi mới này, như đã thấy trong Stable Diffusion, đã góp phần tạo ra những hình ảnh trông như thật

Năm 2017, Google giới thiệu kiến trúc Transformer, một bước đột phá trong xử lý ngôn ngữ tự nhiên. Transformer mã hóa từng từ dưới dạng token, tạo ra bản đồ chú ý giúp nắm bắt mối quan hệ giữa các token. Sự chú ý đến ngữ cảnh này giúp nâng cao khả năng tạo văn bản mạch lạc của mô hình, được minh họa bằng các mô hình ngôn ngữ lớn như ChatGPT

Sự bùng nổ AI tổng quát có được động lực không chỉ nhờ vào các bộ dữ liệu lớn hơn mà còn nhờ những tiến bộ nghiên cứu đa dạng. Những cách tiếp cận này bao gồm GAN, mô hình khuếch tán và transformer, cho thấy phạm vi rộng của các phương pháp góp phần vào lĩnh vực thú vị của AI tạo ra.

## Enter LLMs

Thuật ngữ "Large" trong LLM đề cập đến quy mô tuyệt đối cảu các mô hình này, cả về quy mô kiến trúc lẫn lượng dữ liệu khổng lồ mà chúng được đào tạo. Kích thước quan trọng vì nó cho phép họ nắm bắt được các mô hình và mối quan hệ phức tạp hơn trong ngôn ngữ. Các LLM phổ biến như GPT-3, Gemini, Claude, ... có hàng nghìn tỷ tham số mô hình. Trong bối cảnh học máy, các tham số giống như các nút bấm công tắc của thuật toán điều chỉnh trong quá trình đào tạo để đưa ra dự đoán chính xác hoặc tạo ra kết quả đầu ra có ý nghĩa.

Bây giờ, hãy phân tích ý nghĩa của "Mô hình ngôn ngữ" trong ngữ cảnh này. Mô hình ngôn ngữ về cơ bản là các thuật toán hoặc hệ thống được đào tạo để hiểu và tạo ra văn bản giống con người. Chúng đóng vai trò đại diện cho cách hoạt động của ngôn ngữ, học hỏi từ các bộ dữ liệu đa dạng để dự đoán những từ hoặc chuỗi từ nào có thể xuất hiện tiếp theo trong một ngữ cảnh nhất định.

Khía cạnh "Large" khuếch đại khả năng của họ. Các mô hình ngôn ngữ truyền thống, đặc biệt là các mô hình trước đây có quy mô nhỏ hơn và không thể nắm bắt được sự phức tạp của ngôn ngữ một cách hiệu quả. Với những tiến bộ trong công nghệ và sự sẵn có của sức mạnh tính toán khổng lồ, chúng tôi đã có thể xây dựng những mô hình lớn hơn nhiều. Các mô hình ngôn ngữ lớn này như ChatGPT, có hàng tỷ tham số về cơ bản là các biến mà mô hình sử dụng để hiểu ngôn ngữ.

## Training LLM

Đào tạo LLM là một quá trình phức tạp bao gồm việc hướng dẫn mô hình hiểu và tạo ra văn bản giống con người. Đây là bảng phân tích đơn giản về cách hoạt động của đạo tạo LLM:

<ol>
    <li>Cung cấp văn bản đầu vào: LLM ban đầu được tiếp xúc với dữ liệu văn bản phong phú bao gồm nhiều nguồn khác nhau như sách, bài báo và trang web. Nhiệm vụ của mô hình trong quá trình đào tạo là dự đoán từ hoặc token tiếp theo trình tự dựa trên ngữa cảnh được cung cấp. Nó tìm hiểu các mẫu và mối quan hệ trong dữ liệu văn bản 
    </li>
    <li>Tối ưu hóa trọng lượng mô hình: Mô hình bao gồm các trọng số khác nhau được liên kết với các tham số của nó, phản ảnh tầm quan trọng của các tính năng khác nhau. Trong quá trình đào tạo, các trọng số này sẽ được tinh chỉnh để giảm thiểu tỷ lệ lỗi. Mục tiêu là nâng cao độ chính xác của mô hình trong việc dự đoán từ tiếp theo.
    </li>
    <li> Tinh chỉnh (Fine-tuning) các tham số: LLM liên tục điều chỉnh các giá trị tham số dựa trên phản hồi lỗi nhận được trong quá trình dự đoán. Mô hình tinh chỉnh khả năng nắm bắt ngôn ngữ của nó bằng cách điều chỉnh lặp đi lặp lại các tham số, cải thiện độ chính xác trong việc dự đoán các token tiếp theo.
    </li>
</ol>

Quá trình đào tạo có thể khác nhau tùy thuộc vào loại LLM cụ thể đang được phát triển, chẳng hạn như những loại được tối ưu hóa cho văn bản hoặc hội thoại liên tục. 

Hiệu suất LLM bị ảnh hưởng bởi hai yếu tố chính:
- Kiến trúc mô hình: Thiết kế và độ phức tạp của kiến trúc LLM ảnh hưởng đến khả năng nắm bắt các sắc thái ngôn ngữ của nó.
- Tập dữ liệu: Chất lượng và tính đa dạng của tập dữ liệu được sử dụng để đào tạo là rất quan trọng trong việc hình thành khả năng hiểu ngôn ngữ của mô hình

Đào tạo LLM tư nhân đòi hỏi nguồn lực tính toán và chuyên môn đáng kể. Thời gian của quá trình có thể dao động từ vài ngày đến vài tuần, tùy thuộc vào độ phức tạp của mô hình và kích thước tập dữ liệu. Thông thường, các giải pháp dựa trên đám mây và GPU hiệu suất cao được sử dụng để đẩy nhanh quá trình đào tạo, giúp quá trình đào tạo trở nên hiệu quả hơn. Nhìn chung, đào tạo LLM là một công việc tỉ mỉ và tốn nhiều tài nguyên, đặt nền tảng cho khả năng hiểu và tạo ngôn ngữ của mô hình. 

Sau quá trình đào tạo ban đầu, LLM có thể được tùy chỉnh dễ dàng cho các nhiệm vụ khác nhau bằng cách sử dụng các bộ dữ liệu được giám sát tương đối nhỏ, mộ quy trình được gọi là tinh chỉnh.  

Có 3 mô hình học tập phổ biến:
- __Zero-shot learning__: LLM cơ sở có thể xử lý nhiều loại yêu cầu mà không cần đào tạo rõ ràng, thường bằng cách sử dụng lời nhắc, mặc dù độ chính xác của phản hồi có thể khác nhau.
- __Few-shot learning__: Bằng cách cung cấp một số lượng nhỏ các ví dụ đào tạo tích hợp hiệu suất của mô hình cơ sở sẽ cải thiện đáng kể trong một miền cụ thể.
- __Domain Adaptation__: Điều này mở rộng từ việc học vài lần trong đó những người thực hành huấn luyện một mô hinnhf cơ sở để điều chỉnh các tham số của nó bằng cách sử dụng dữ liệu bổ sung có liên quan đến ứng dụng hoặc miền cụ thể.

## LLM Real World Usecases

![llm-usecase](./img/llm_usecase.png)

1. __Tạo nội dung__: LLM vượt trội trong việc tạo nội dung bằng cách hiểu ngữ cảnh và tạo ra văn bản mạch lạc và phù hợp với ngữ cảnh. Chúng có thể được sử dụng để tự động tạo nội dung sáng tạo cho hoạt động tiếp thị, bài đăng trên mạng xã hội và các tìa liệu truyền thông khác, đảm bảo chất lượng và mức độ liên quan cao. Một vài ứng dụng trong thế giới thực: Nền tảng tiếp thị, công cụ quản lý phương tiện, truyền thông xã hội, nền tảng tạo nội dung, đại lý quảng cáo. 
2. __Dịch ngôn ngữ__: LLM có thể cải thiện đáng kể nhiệm vụ dịch ngôn ngữ bằng cách hiểu các sắc thái của các ngôn ngữ khác nhau. Chúng có thể cung cấp các bản dịch chính xác và phù hợp với ngữ cảnh, khiến chúng trở thành công cụ có giá trị cho các doanh nghiệp hoạt động trong môi trường đa ngôn ngữ. Điều này có thể tăng cường truyền thông và tiếp cận toàn cầu. Một vài ứng dụng trong thế giới thực: Dịch vụ dịch, nền tảng giao tiếp toàn cầu, ứng dụng kinh doanh quốc tế.
3. __Tóm tắt văn bản__: LLM rất thành thạo trong việc tóm tắt các tài liệu dài bằng các định thông tin chính và duy trì thông điệp cốt lõi. Khả năng này rất có giá trị đối với những người sáng tạo nội dung, nhà nghiên cứu và doanh nghiệp muốn nhanh chóng trích xuất những hiểu biết cần thiết từ khối lượng lớn văn bản, nâng cao hiệu quả trong việc tiêu. Một vài ứng dụng trong thế giới thực: Công cụ nghiên cứu, công cụ tổng hợp tin tức, nền tảng quản lý nội dung.
4. __Kiểm duyệt nội dung__: LLM có thể được sử dụng để kiểm duyệt nội dung bằng cách phân tích văn bản và xác định nội dung có thể không phù hợp hoặc có hại, điều này giúp duy trì môi trường trực tuyến an toàn và tôn trọng bằng cách tự động gắn cờ hoặc lọc ra nội dung vi phạm nguyên tắc, đảm bảo an toàn cho người dùng. Một vài ứng dụng trong thế giới thực: Nền tảng truyền thông xã hội, diễn đàn trực tuyến, công cụ quản lý cộng đồng
5. __Truy xuất thông tin__: LLM có thể nâng cao hệ thống truy xuất thông tin bằng cách hiểu các truy vấn của người dùng và truy xuất thông tin liên quan từ các bộ dữ liệu lớn. Điều này đặc biệt hữu ích trong các công cụ tìm kiếm, cơ sở dữ liệu và hệ thống quản lý tri thức, nơi LLM có thể cải thiện độ chính xác của kết quả tìm kiếm. Một vài ứng dụng trong thế giới thực: Công cụ tìm kiếm, hệ thống cơ sở dữ liệu, nền tảng quản lý tri thức
6. __Trả lời câu hỏi và chatbot__: LLM có thể được sử dụng cho các nhiệm vụ trả lời câu hỏi, trong đó chúng hiểu ngữ cảnh của câu hỏi và tạo ra các câu trả lời chính xác và phù hợp. Chúng cho phép các hệ thống này tham gia vào các cuộc hội thoại tự nhiên hơn và phù hợp với ngữ cảnh hơn, hiểu được các truy vấn của người dùng và đưa ra các phản phù hợp. Một vài ứng dụng trong thế giới thực: Hệ thống chăm sóc KH, chatbot, trợ lý ảo và nền tảng giáo dục
7. **Công cụ giáo dục**: LLM đóng góp cho các công cụ giáo dục bằng cách cung cấp giao diện ngôn ngữ tự nhiên cho nền tảng học tập. Họ có thể hỗ trợ sinh viên tạo ra các bản tóm tắt, trả lời các câu hỏi và tham gia vào các cuộc trờ chuyện học tập mang tính thương tác. Điều này tạo điều kiện cho trải nghiệm học tập được các nhân hóa và hiệu quả.

## LLM Challenges
Các thách thức trong LLM được phân loại thành các chủ đề khác nhau:
- **Thách thức về dữ liệu**: Điều này liên quan đến dữ liệu được sử dụng để đào tạo và cách mô hình giải quyết các khoảng trống hoặc dữ liệu bị thiếu.
  
   *Xu hướng dữ liệu*: Sự hiện diện của những thành kiến và sự mất cân bằng trong dữ liệu huấn luyện dẫn đến kết quả đầu ra của mô hình bị sai lệch.

  *Kiến thức và ảo giác về thế giới hạn chế*: LLM có thể thiếu hiểu biết toàn diện về các sự kiện và thông tin trong thế giới thực và có xu hướng tạo ra thông tin ảo giác. Lưu ý rằng việc đào tạo họ về dữ liệu mới là một quá trình dài và tốn kém.

  *Phụ thuộc vào chất lượng dữ liệu đào tạo*: Hiệu suất LLM bị ảnh hưởng nặng nề bởi chất lượng và tính đại diện của dữ liệu đào tạo.
  
- **Những thách thức về đạo đức, xã hội**: Điều này liên quan đến việc giải quyết các vấn đề như giảm thiểu thành kiến, đảm bảo quyền riêng tư và ngăn chặn việc tạo ra nội dung có hại trong quá trình triển khai LLM.

   *Mối quan tâm về đạo đức* là những mối quan tâm liên quan đến việc sử dụng các mô hình ngôn ngữ một cách có trách nhiệm và có đạo đức, đặc biệt là trong các bối cảnh nhạy cảm.

  *Khuếch đại sai lệch* là những sai lệch có trong dữ liệu huấn luyện có thể trở nên trầm trọng hơn, dẫn đến kết quả đầu ra không công bằng hoặc mang tính phân biệt đối xử.

  *Các vấn đề pháp lý và bản quyền* là các vấn đề tiềm ẩn phát sinh từ nội dung được ra vi phạm bản quyền hoặc vi phạm pháp luật.

  *Mối lo ngại về quyền riêng tư của người dùng*: rủi ro liên quan đến việc tạo văn bản dựa trên thông tin đầu vào của người dùng, đặc biệt khi xử lý thông tin riêng tư hoặc nhạy cảm.
  
- **Thách thức về kỹ thuật**: Tập trung về việc thực thi.

  *Tài nguyên tính toán*: Cần có sức mạnh tính toán đáng kể để đào tạo và triển khai các mô hình ngôn ngữ lớn.

  *Khả năng diễn giải*: Những thách thức trong việc hiểu và giải thích quá trình ra quyết định của các mô hình phức tạp.

  *Đánh giá*: Đưa ra một thách thức đáng chú ý vì việc đánh giá các mô hình nhiều nhiệm vụ và lĩnh vực khác nhau được thiết kế không phù hợp, đặc biệt là do những thách thức do nội dung được tạo tự do đặt ra.

  *Tinh chỉnh thử thách* là khó khăn trong việc điều chỉnh các mô hình được đào tạo trước cho phù hợp với các nhiệm vụ hoặc lĩnh vực cụ thể.

  *Hiểu về bối cảnh*: LLM có thể đối mặt với những thách thức trong việc duy trì bối cảnh mạch lạc trong các đoạn văn hoặc cuộc hội thoại dài hơn.

  *Tính mạnh mẽ trước các cuộc tấn công bất lợi*: Dễ bị tổn thương trước các hình vi cố ý thao túng dữ liệu đầu vào dẫn đến kết quả đầu ra không chính xác.

  *Bối cảnh dài hạn*: gặp khó khăn trong việc duy trì bối cảnh và sự mạch lạc trong các đoạn văn bản hoặc cuộc thỏa luận mở rộng
  
- **Thách thức về triển khai**: Quan tâm đến các quy trình cụ thể liên quan đến việc chuyển đổi LLM đầy đủ chức năng sang các trường hợp sử dụng trong thực tế.

  *Khả năng mở rộng*: đảm bảo rằng mô hình có thể mở rộng quy mô một cách hiệu quả để xử lý khối lượng công việc và nhu cầu ngày càng tăng trong môi trường production.

  *Độ trễ*: Giảm thiểu thời gian phản hồi hoặc độ trễ của mô hình để cung cấp các tương tác nhanh chóng và hiểu quả, đặc beietj là trong các ứng dụng thời gian thực.

  *Giám sát và bảo trì*: Triển khai các hệ thống giám sát mạnh mẽ để theo dõi hiệu suất của mô hình, phát hiện sự cố và thực hiện maintain thường xuyên để tránh thời gian ngừng hoạt động.

  *Tích hợp với những hệ thống hiện có*: đảm bảo việc tích hợp mượt mà off LLMs cùng với phần mềm, cơ sở dữ liệu và hạ tầng trong một tổ chức.

  *Quản lý chi phí* là tối ưu chi phí triển khai và duy trì LLMs, vì chúng có thể tốn nhiều tài nguyên cả về tính toán và lưu trữ.

  *Mối quan tâm về bảo mật* là giải quyết các lỗ hổng bảo mật tiềm ẩn và rủi ro liên quan đến việc triển khai các mô hình ngôn ngữ trong production, bao gồm cả việc bảo vệ khỏi các cuộc tấn công độc hại.

  *Khả năng tương tác* đảm bảo khả năng tương thích với các công cụ, khung hoặc hệ thống khác có thể là một phần của quy trình production tổng thể

  *Kết hợp phản hồi của người dùng* phát treieenr các cơ chế kết hợp phản hồi của người dùng đẻ liên tục cải tiến và cập nhật mô hình trong môi trường production.

  *Tuân thủ quy định* là tuân thủ các yêu cầu quy định và tiêu chuẩn tuân thủ, đặc biệt là trong các ngành có quy định nghiêm ngặt về bảo vệ dữ liệu và quyền riêng tư

  *Xử lý nội dung động* là quản lý việc tạo văn bản trong môi trường động nơi nội dung và tương tác của người dùng thay đổi thường xuyên
